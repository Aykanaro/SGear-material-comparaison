```python
# Reading
path = [
    './data/materials/aluminum.json',
    './data/materials/crimson_iron.json',
    './data/materials/diamond.json'
    ]

data = []
for p in path:
    with open(p) as folder:
        data.append(json.load(folder))
data[2]
```

```python
# Stats
stats_main = pd.DataFrame(columns=list(data[2]['stats']['main'].keys()))
pb = []

path = './data/materials/'
arr = os.listdir(path)

for a in arr:
    with open(path + a) as folder:
        f = json.load(folder)
        try:
            stats_main.loc[a[:-5], list(f['stats']['main'].keys())] = f['stats']['main']
        except:
            pb.append(a[:-5])
```
